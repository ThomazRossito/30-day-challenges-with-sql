# Solution Challenge 7 - Thomaz Antonio Rossito Neto

````python
%%sql
create table Track2 
AS 
    select tr.TrackId                 as TrackId,
           at.Name                    as artista,
           tr.Name                    as musica,
           CASE 
                WHEN tr.Composer is null THEN ''      
           ELSE tr.Composer
           END                        as compositor,
           ge.Name                    as genero,
           CAST(tr.bytes AS INT)/1024 as size_mb
      from track as tr
     inner join genre as ge
        on tr.GenreId = ge.GenreId    
     inner join album as al
        on tr.AlbumId = al.AlbumId
     inner join Artist as at
        on al.ArtistId = at.ArtistId;
````

````python
%%sql
select *
  from Track2
````
